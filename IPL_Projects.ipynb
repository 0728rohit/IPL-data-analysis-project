```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("matches.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Deleting the 'method' column since it has so many NaN values
df = df.drop('method', axis=1)
```

```python
df.head()
```

```python
df['winner']=df['winner'].fillna('Unknown')
```

```python
df['city']=df['city'].fillna('Unknown')
```

```python
df['result_margin']=df['result_margin'].fillna(0)


df['result_margin'] = df['result_margin'].interpolate(method='linear')
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['target_runs']=df['target_runs'].fillna(0)


df['target_runs'] = df['target_runs'].interpolate(method='linear')
```

```python
df['target_overs']=df['target_overs'].fillna(0)


df['target_overs']= df['target_overs'].interpolate(method='linear')
df['target_runs']=df['target_runs'].astype(int)

df['result_margin']=df['result_margin'].astype(int)
df.isnull().sum()
```

```python
df['super_over'] = df['super_over'].replace('N', 'No')
df['super_over'] = df['super_over'].replace('Y', 'Yes')
df.head()
df.isnull().sum()
```

```python
df['player_of_match']=df['player_of_match'].fillna('Unknown')
df.head(20)
df.isnull().sum()
```

```python
df.head()
```

```python
season_counts = df['season'].value_counts()
season_counts
```

```python
plt.figure(figsize=(12,6))
plt.bar(season_counts.index,season_counts.values)
plt.title('Number of Matches per Season',color="blue")
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.yticks(np.arange(0,110,10),rotation=90)
plt.show()

```

```python
highest_runs_by_each_team=df.groupby("team1")["target_runs"].max()
print("Highest score by each team(2008-2024)")
highest_runs_by_each_team
```

```python
plt.figure(figsize=(15,10))
colors=["yellow","lightblue","orange","violet","orange","purple","darkblue","blue","peachpuff","red","pink","thistle","lightcoral","lightsalmon"]
plt.barh(highest_runs_by_each_team.index,highest_runs_by_each_team.values,color=colors)
plt.title("Highest score by each team(2008-2024)")
plt.xlabel("Runs")
plt.ylabel("Teams")
plt.xticks(np.arange(0,310,10),rotation=360)
plt.show()
```

```python
df['toss_decision']
```

```python

wins_after_winning_toss = df[df['winner'] == df['toss_winner']]


wins_after_toss_counts = wins_after_winning_toss['winner'].value_counts()


wins_after_toss_counts
```

```python
loss_after_loosing_toss = df[df['winner'] != df['toss_winner']]

loss_after_toss_counts = loss_after_loosing_toss['winner'].value_counts()

print("Wins After Lossing Toss")
loss_after_toss_counts
```

```python
# Calculating the total wins after winning the toss
total_wins_after_toss_win = wins_after_toss_counts.sum()

# Calculating the total wins after losing the toss
total_wins_after_toss_loss = loss_after_toss_counts.sum()

# Data for the pie chart
sizes = [total_wins_after_toss_win, total_wins_after_toss_loss]
labels = ['Wins after winning toss', 'Wins after losing toss']
colors = ['lightgreen', 'salmon']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Proportion of Matches Won Based on Toss Outcome')
plt.show()
```

```python
# Head To head Between Kkr and Rcb
kkr_rcb_matches = df[((df['team1'] == "Kolkata Knight Riders") & (df['team2'] == 'Royal Challengers Bangalore')) |
                    ((df['team1'] == 'Kolkata Knight Riders') & (df['team2'] == 'Royal Challengers Bangalore'))]

kkr_rcb_wins = kkr_rcb_matches['winner'].value_counts()
print("Head-to-head wins between Kolkata Knight Riders and Royal Challengers Banglore:")
kkr_rcb_wins
```

```python
plt.figure(figsize=(10,5))
colors=["purple","red"]
plt.barh(kkr_rcb_wins.index,kkr_rcb_wins.values,color=colors)
plt.title("Head-to-head wins between Kolkata Knight Riders and Royal Challengers Banglore")
plt.xlabel("Teams")
plt.ylabel("Number of wins")
plt.legend()
plt.show()
```
